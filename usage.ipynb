```python
# rules = {"stock_num":200,                            # pick exact 200 stocks
#          "tcost":0.005 * 1,                          # tcost 
#          "section":[sect_constrains, 0., True],    # expect value, tolerance
#          "mv":[0.32, 0.02, True],                    # expect value, tolerance
#          "trend":[0.05, 0.02, False]}   

for n, date in enumerate(SIM_DATES):
    control = 0.035
    delta = (0.08-control)/(len(SIM_DATES) - 30)
    rules = {"stock_num":200,                         
         "tcost":0.005 * 1,                         
         "section":[sect_constrains, max(0.08 -  n*delta, control), True],
         "mv":[0.32, 0.03, True],                  
         "trend":[0.05, 0.02, False]}   
    
    print("\n"*2,"-"*30,date, round(max(0.08 -  n*delta, control), 3),"-"*30)
    stocks = genInputSimple(date)
    pool = Portfolio(stocks)
    solver = Solver(pool, rules=rules)
    %time solver.solve()
    save_path = os.path.join(foldpath, "weight", "Weight.{}.csv".format(date))
    solver.evaluate(save_as = save_path)
```
